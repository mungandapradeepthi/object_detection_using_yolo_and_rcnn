```python
!git clone https://github.com/DatumLearning/Faster_RCNN_Opencv
#for accuracy
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Load Faster R-CNN
weights="/content/Faster_RCNN_Opencv/frozen_inference_graph.pb"
config="/content/Faster_RCNN_Opencv/mask_rcnn_inception_v2_coco_2018_01_28.pbtxt"
net=cv2.dnn.readNetFromTensorflow(weights,config)

# Load COCO class names
coco_names = []
with open("/content/Faster_RCNN_Opencv/coco.names", "r") as f:
    coco_names = [line.strip() for line in f.readlines()]

# Load input image
image=cv2.imread("/content/Faster_RCNN_Opencv/traffic4.jpg")
height,width,_=image.shape

# Preprocess the image
blob=cv2.dnn.blobFromImage(image,swapRB=True,crop=False)

# Set the input to the network
net.setInput(blob)

# Run the forward pass to get the output
boxes=net.forward(["detection_out_final"])
output=boxes[0].squeeze()

# Count the number of objects detected for each class
objects_count = {}
for i in range(output.shape[0]):
  conf=output[i,2]
  if conf>0.8:
    class_id=int(output[i,1])
    if class_id in objects_count:
      objects_count[class_id] += 1
    else:
      objects_count[class_id] = 1

# Create a bar plot of object detection accuracy
x = np.arange(len(objects_count))
y = list(objects_count.values())
labels = list(objects_count.keys())

plt.bar(x, y)
plt.xlabel('Class')
plt.ylabel('Number of Objects Detected')
plt.xticks(x, labels)
plt.title('Accuracy of Object Detection using Faster R-CNN')
plt.show()

step-3:for precision and recall


import cv2
import numpy as np
from sklearn.metrics import precision_score, recall_score
from google.colab.patches import cv2_imshow

# Load Faster R-CNN
weights="/content/Faster_RCNN_Opencv/frozen_inference_graph.pb"
config="/content/Faster_RCNN_Opencv/mask_rcnn_inception_v2_coco_2018_01_28.pbtxt"
net=cv2.dnn.readNetFromTensorflow(weights,config)

# Load COCO class names
coco_names = []
with open("/content/Faster_RCNN_Opencv/coco.names", "r") as f:
    coco_names = [line.strip() for line in f.readlines()]

# Load input image
image=cv2.imread("/content/Faster_RCNN_Opencv/traffic4.jpg")
height,width,_=image.shape

# Preprocess the image
blob=cv2.dnn.blobFromImage(image,swapRB=True,crop=False)

# Set the input to the network
net.setInput(blob)

# Run the forward pass to get the output
boxes=net.forward(["detection_out_final"])
output=boxes[0].squeeze()

# Create ground truth labels for traffic signs
ground_truth = []
for i in range(output.shape[0]):
  conf=output[i,2]
  if conf>0.8:
    x1n,y1n,x2n,y2n=output[i,3:]
    x1=int(x1n*width)
    y1=int(y1n*height)
    x2=int(x2n*width)
    y2=int(y2n*height)
    class_id=int(output[i,1])
    ground_truth.append([x1,y1,x2,y2,class_id])

# Get predictions from the output
predictions = []
for i in range(output.shape[0]):
  conf=output[i,2]
  if conf>0.8:
    x1n,y1n,x2n,y2n=output[i,3:]
    x1=int(x1n*width)
    y1=int(y1n*height)
    x2=int(x2n*width)
    y2=int(y2n*height)
    class_id=int(output[i,1])
    predictions.append([x1,y1,x2,y2,class_id])

# Calculate precision and recall scores for each class
precision_scores = []
recall_scores = []
for i in range(len(coco_names)):
    true_positives = len([p for p in predictions if p[4] == i])
    false_positives = len([p for p in predictions if p[4] != i])
    false_negatives = len([p for p in ground_truth if p[4] != i])

    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0

    precision_scores.append(precision)
    recall_scores.append(recall)

# Calculate the average precision and recall scores
average_precision = np.mean(precision_scores)
average_recall = np.mean(recall_scores)

print(f"Average Precision: {average_precision}")
print(f"Average Recall: {average_recall}")
```
